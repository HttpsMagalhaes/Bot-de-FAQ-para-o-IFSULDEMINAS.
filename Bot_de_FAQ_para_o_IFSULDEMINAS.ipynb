```python
#transformando a lista (faqs) em dataframe para acesso dos mesmos
df = pd.DataFrame(faqs)
df.columns = ["Titulo", "Conteudo"]
df
```

```python
#modelo utilizado
model = "models/embedding-001"
```

```python
#função que cria o embedding
def embed(title, text):
  return genai.embed_content(model = model,
                                 content = text,
                                 title = title,
                                 task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
#adiciona a coluna embedding no dataframe
df["Embeddings"] = df.apply(lambda row: embed(row["Titulo"], row["Conteudo"]), axis=1)
df
```

```python
#função de retorno para dúvida
def faq_query(query, base, model):
  query_embedding = genai.embed_content(model = model,
                                 content = query,
                                 task_type="RETRIEVAL_QUERY")["embedding"]

  comparison = np.dot(np.stack(df["Embeddings"]), query_embedding)

  indice = np.argmax(comparison)
  return df.iloc[indice]["Conteudo"]
```

```python
#testes
query = "preciso do acesso a minha carteira"

response = faq_query(query, df, model)
print(response)
```
